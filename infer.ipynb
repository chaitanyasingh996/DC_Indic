```python
import json
import os

def convert_to_via_format(ocr_dict, filename, size=0):
    """
    Convert OCR dictionary to VIA annotation format.
    """
    regions = []

    for key, value in ocr_dict.items():
        text = value["txt"]
        script = value["script_language"]
        x1, y1, x2, y2 = value["bbox"]

        region = {
            "shape_attributes": {
                "name": "rect",
                "x": x1,
                "y": y1,
                "width": x2 - x1,
                "height": y2 - y1
            },
            "region_attributes": {
                "script_language": script,
                "text": text
            }
        }

        regions.append(region)

    via_key = f"{filename}{size}"

    return {
        via_key: {
            "filename": filename,
            "size": size,
            "regions": regions,
            "file_attributes": {}
        }
    }


# ------------------------------------------------------------------
#  Function to merge new OCR result into output.json
# ------------------------------------------------------------------
def update_output_json(new_data, output_file="output.json"):
    # Step 1: Load existing JSON (if file exists)
    if os.path.exists(output_file):
        with open(output_file, "r", encoding="utf-8") as f:
            try:
                existing = json.load(f)
            except:
                existing = {}   # in case file is empty or corrupted
    else:
        existing = {}

    # Step 2: Update existing JSON with new VIA data
    existing.update(new_data)

    # Step 3: Overwrite output.json with full updated data
    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(existing, f, ensure_ascii=False, indent=4)

    print(f"Updated {output_file}")
```

```python
for i in range(31,83):
    if i >= 51:
        filename = "dc_"+str(i)+".png"
    else:
        filename = "dc_"+str(i)+".jpg"
    if i == 34 or i == 70:
        continue;
    path = "test_images\Raw_Images\\" + filename
    k = ocr_system.ocr(path)

    size = os.path.getsize(path) if os.path.exists(path) else 0

    via_data = convert_to_via_format(k, filename, size)

    # Add this to output.json
    update_output_json(via_data)

    print("Saved output.json")
```
